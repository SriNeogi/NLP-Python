```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv('Books.csv')
df2 = pd.read_csv('Users.csv')
df3 = pd.read_csv('Ratings.csv')
```

```python
#Merge Ratings and Books, based on a book's ISBN
df4=pd.merge(df1,df3, on='ISBN')
df4.head()
```

```python
#Merge Ratings and Books and Users, based on User-ID
df5=pd.merge(df4,df2, on='User-ID')
#Drop IMAGE columns
df5.drop(['Image-URL-S', 'Image-URL-M', 'Image-URL-L'], axis=1, inplace=True)
#Final df
df5.head()
df5.shape
```

```python
#Find the book with most number of ratings
df5.groupby('Book-Title')['Book-Rating'].count().sort_values(ascending=False).head()
```

```python
#Average rating of each book. Create a ratings DF
ratings = pd.DataFrame(df5.groupby('Book-Title')['Book-Rating'].mean())
ratings.head()
```

```python
#Average rating+Num of ratings DF
ratings['num of ratings'] = pd.DataFrame(df5.groupby('Book-Title')['Book-Rating'].count())
ratings.sort_values('num of ratings',ascending=False).head(20)
```

# Choosing top 5 authors

```python
#This dataset is quite large, so we'll work with a smaller dataset
#Find out the top 5 authors
top5authors=pd.DataFrame(df5['Book-Author'].value_counts().head(5))
top5authors
```

```python
rowlabels = top5authors.index.values
rowlabels
```

# Data Subset based on top 5 authors

```python
#Smaller df
df6=df5[df5["Book-Author"].isin(['Stephen King', 'Nora Roberts', 'John Grisham', 'James Patterson',
       'Mary Higgins Clark'])]
df=df6[['User-ID','Book-Title','Book-Rating']]
df.shape
```

```python
#Creating a pivot table for pandas series correlation
bookmat = df.pivot_table(index='User-ID',columns='Book-Title',values='Book-Rating')
bookmat.tail()
```

# Find books similar to THE PELICAN BRIEF BY JOHN GRISHAM

```python
#CORRWITH FUNCTION
pelicanbrief_user_ratings = bookmat['The Pelican Brief']
similar_to_pelicanbrief = bookmat.corrwith(pelicanbrief_user_ratings)
similar_to_pelicanbrief
```

```python
corr_pelicanbrief = pd.DataFrame(similar_to_pelicanbrief,columns=['Correlation'])
corr_pelicanbrief.dropna(inplace=True)
corr_pelicanbrief.head()
```

```python
corr_pelicanbrief.sort_values('Correlation',ascending=False).head(10)
```

```python
df_new=pd.merge(ratings,corr_pelicanbrief, on='Book-Title')
df_new.sort_values(by=['Correlation', 'num of ratings_y'],ascending=False).head(20)
```

# Conclusion

The recommender system suggests that people who like 'The Pelican Brief' by John Grisham will also like books by Nora Roberts and Stephen King. Other books highly correlated with 'The Pelican Brief' are 'A Time to Kill' and 'The Runaway Jury'. Both are Grisham's works.
